# Generating Signal

```python
import numpy as np
from scipy import fftpack
import matplotlib.pyplot as plt

time_step = 0.05
time_vec = np.arange(0,40,time_step)
period = 5
sig = np.sin(2*np.pi*time_vec/period) + 0.025*np.random.randn(time_vec.size)
# print(np.round(sig,2))

sig_fft = fftpack.fft(sig)
plt.figure(figsize =(13,4))
plt.plot(sig_fft)
```

# Fourier Transform

```python
import numpy as np
from scipy import fftpack
import matplotlib.pyplot as plt
y_1 = np.arange(0,5,1)
c = fftpack.fft(y_1)

plt.plot(abs(c), "teal")

y_2 = fftpack.ifft(c)

plt.plot(np.real(y_2), "orange")
```

# Fourier transform of sinusoidal wave with added noise

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import fftpack


time_vec = np.arange(0, 5, 0.05)
sig = (np.sin(2*np.pi*time_vec/5)) 
noise = 0.15*np.random.randn(time_vec.size)
noise_sig = sig + noise
in_freq = fftpack.fftfreq(noise_sig.size,0.05)

plt.figure(figsize =(16,10))
plt.subplot(2,2,1)
plt.title("Time")
plt.plot(time_vec)
plt.subplot(2,2,2)
plt.title("Signal")
plt.plot(sig)
plt.subplot(2,2,4)
plt.title("Signal With Noise")
plt.plot(noise_sig)
plt.subplot(2,2,3)
plt.title("Fourier Transform")
plt.plot(in_freq)

plt.subplot(2,2,4)
plt.plot(sig)
```
