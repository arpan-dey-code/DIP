# Plotting Using Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt

t=np.arange(0,1,0.01)

xt = np.sin(2*np.pi*5*t)

plt.plot(t,xt)
plt.title("Wave")
plt.xlabel("Time")
plt.ylabel("Amplitude")
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.arange(0,1,0.01)
f = 5
xt = np.sin(2*np.pi*f*t)

plt.stem(t,xt)
```

```python
import matplotlib.pyplot as plt
n = [0,1,2,3,4,5,6]
xn = [-1,4,-3,0,9,8,9]
plt.stem(n,xn)
```

```python
import numpy as np
import matplotlib.pyplot as plt
t = np.arange(0,1,0.01)
xt1 = np.cos(2*np.pi*3*t)
xt2 = np.cos(2*np.pi*5*t)
xt3 = np.cos(2*np.pi*7*t)
xt4 = np.cos(2*np.pi*9*t)

plt.subplot(2,2,1) 
plt.plot(t,xt1)
plt.subplot(2,2,2) 
plt.plot(t,xt2)
plt.subplot(2,2,3) 
plt.plot(t,xt3)
plt.subplot(2,2,4) 
plt.plot(t,xt4)


```
